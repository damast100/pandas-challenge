```python
# importing dependencies
import pandas as pd
```

```python
# Creating variables for each file path
school = "C:\\Users\\Derek\\Desktop\\pandas-challenge\\Resources\\schools_complete.csv"
student = "C:\\Users\\Derek\\Desktop\\pandas-challenge\\Resources\\students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school)
student_data = pd.read_csv(student)
```

```python
# Merge the two files together
complete_data = pd.merge(student_data, school_data, how="left", on="school_name")
complete_data_df = pd.DataFrame(complete_data)
complete_data_df.head()
```

```python
# total number of schools
number_of_schools = len(complete_data_df["school_name"].unique())
number_of_schools
```

```python
# total number of students
number_of_students = len(complete_data_df["Student ID"].unique())
number_of_students
```

```python
# total budget
budget = complete_data_df["budget"].unique()
total_budget = budget.sum()
total_budget
```

```python
# average math score
math = complete_data_df["math_score"].sum()
math_average = math / number_of_students
math_average
```

```python
# average reading score
reading = complete_data_df["reading_score"].sum()
reading_average = reading / number_of_students
reading_average
```

```python
# Percentage of students passing math
passing_math_df = complete_data_df.loc[complete_data_df["math_score"] >= 70]
number_passing_math = passing_math_df["Student ID"].count()
math_percent_passing = (number_passing_math / number_of_students) * 100
math_percent_passing
```

```python
# Percentage of students passing reading
passing_reading_df = complete_data_df.loc[complete_data_df["reading_score"] >= 70]
number_passing_reading = passing_reading_df["Student ID"].count()
reading_percent_passing = (number_passing_reading / number_of_students) * 100
reading_percent_passing
```

```python
# Percentage of students passing both math and reading
passing_both_df = complete_data_df.loc[(complete_data_df["math_score"] >= 70) & (complete_data_df["reading_score"] >= 70)]
number_passing_both = passing_both_df["Student ID"].count()
passing_both_percent = (number_passing_both / number_of_students) * 100
passing_both_percent
```

```python
# creating the district summary dataframe
district_summary = {"Total Schools": [number_of_schools],
                    "Total Students": [number_of_students],
                    "Total Budget": [total_budget],
                    "Average Math Score": [math_average],
                    "Average Reading Score": [reading_average],
                    "% Passing Math": [math_percent_passing],
                    "% Passing Reading": [reading_percent_passing],
                    "Overall Passing": [passing_both_percent]}

district_summary_df = pd.DataFrame(district_summary)                     
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```
